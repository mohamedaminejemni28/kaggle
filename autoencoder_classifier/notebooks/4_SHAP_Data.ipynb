```python
"""
Created on Thu Jul  4 15:08:48 2024

@author: roqui
"""
import matplotlib.pyplot as plt  # for making plots / graphs
import pandas as pd              # for reading the .csv file and related operations
import numpy as np               # for working with arrays (multi-dimensional)
from   sklearn.preprocessing   import StandardScaler
from   sklearn.model_selection import StratifiedKFold
from   sklearn.svm             import SVC
# Use SHAP module to get Model Interpretability.
import shap
import os

"""
Created on Thu Jul  4 15:08:48 2024

@author: roqui
"""
import matplotlib.pyplot as plt  # for making plots / graphs
import pandas as pd              # for reading the .csv file and related operations
import numpy as np               # for working with arrays (multi-dimensional)
from   sklearn.preprocessing   import StandardScaler
from   sklearn.model_selection import StratifiedKFold
from   sklearn.svm             import SVC
# Use SHAP module to get Model Interpretability.
import shap
import os

#path_folder = f'Invariant Speeds Young and Older'
path_folder = f'Sample Data'
path_train  = f'{path_folder}/ALL FEATURES OLDER - Pes Planus and Control JUNE 25 Train.xlsx'
path_test   = f'{path_folder}/ALL FEATURES OLDER - Pes Planus and Control JUNE 25 Test.xlsx'
path_scores = f'{path_folder}/Scores_RBF_older.xlsx'

# Read the Excel file
EXCEL_train = pd.ExcelFile(path_train)
EXCEL_test = pd.ExcelFile(path_test)
EXCEL_scores = pd.ExcelFile(path_scores)

# Dictionary to hold the data from each sheet
sheets_data   = {}
SHAP_DATA     = {}
Features_SHAP = pd.DataFrame()

for sheet_name in EXCEL_scores.sheet_names:
    # READ TRAIN SET
    # LEER EL DATA SET DE DATOS PRINCIPAL
    df_train      = pd.read_excel(EXCEL_train, sheet_name=sheet_name)
    df_inf_Train  = df_train.iloc[:,0:4]

    df_Scores     = pd.read_excel(EXCEL_scores, sheet_name=sheet_name)

    SHAP_DATA[sheet_name] = {}

    # model parameters
    C_VAL_list      = list(df_Scores['C_Value'].iloc)
    R_VAL_list      = list(df_Scores['Gamma_Value'].iloc)
    CV_SPLIT_list   = list(df_Scores['CV_split'].iloc)
    RS_list         = list(df_Scores['Random_State'].iloc)
    features_list   = list(df_Scores['Features'].iloc)
    features_indx   = list(df_Scores['Feature_idx'].iloc)
    feat_model_name = list(df_Scores['Name_Model'].iloc)
    df_participants = df_train.iloc[:,0:4]

    # THE FIRST AND LEAST FEATURES
    a = 'Pelv_Angle_Y_MAX_SW'
    b = 'Hip_Angle_Z_OHS'

    # DEFINIR EL CAMBIO DE VARIABLE EN LA CLASIFICACIÓN OBJETIVA -y
    df_train["Group"] = df_train["Group"].replace({'Control': 0, 'Flatfoot': 1})

    # Define the different segments from dataset to be used.
    segments_train = {
        'All_Data' : df_train.loc[:, a:b]
    }

    # saving the target variables into `y` variable.
    Y_train = df_train.loc[:, "Group"].values

    # READ TEST SET
    df_test_p   = pd.read_excel(EXCEL_test, sheet_name=sheet_name)
    df_inf_Test = df_test_p.iloc[:,0:4]
    df_test = df_test_p.drop(["Participant", "Index", "Side"], axis=1)
    df_test["Group"] = df_test["Group"].replace({'Control': 0, 'Flatfoot': 1})

    segments_test = {
        'All_Data' : df_test.loc[:, a : b]
    }

    Y_test  = df_test.loc[:, "Group"].values

    # CREATE THE FOLDER WHERE THE SHAP'S PLOTS WILL BE SAVED
    output_dir = f'{path_folder}/SHAPS_TRAIN_and_TEST_Plots'
    os.makedirs(output_dir, exist_ok=True)

    sc = StandardScaler()
    counter = 0

    for features_str in features_list:
        # model parameters
        C_VAL = C_VAL_list[counter]
        R_VAL = R_VAL_list[counter]
        CV_SPLIT     = CV_SPLIT_list[counter]
        RANDOM_STATE = RS_list[counter]

        # GET NAME MODEL
        df_name = f'Model_{feat_model_name[counter]}'

        # TRANSFORM THE DATA ON THE CORRECT FORMAT
        features = features_str.strip("[]").split()
        features = np.array(features)
        sfs_feature_names = [int(feature) for feature in features]

        print("Number of Features selected: ", len(sfs_feature_names))
        print("df.shape = ", segments_train['All_Data'].iloc[:, sfs_feature_names].shape)

        segment_train = segments_train['All_Data'].iloc[:, sfs_feature_names]
        segment_test  = segments_test['All_Data'].iloc[:, sfs_feature_names]
        cols = segment_train.columns.values.tolist()
        print(len(cols))

        temp = pd.DataFrame(segments_train["All_Data"].keys().to_numpy(), columns=["FeatureNames"])
        features_value = temp.iloc[sfs_feature_names]

        kfold_cv = StratifiedKFold(n_splits = CV_SPLIT, shuffle=True, random_state = RANDOM_STATE)
        X_train  = segment_train
        X_test   = segment_test

        ix_training, ix_test = [], []
        # Loop through each fold and append the training & test indices to the empty lists above
        for fold in kfold_cv.split(X_train, Y_train):
            ix_training.append(fold[0]), ix_test.append(fold[1])
            print(len(fold[1]))

        len(ix_training)
        SHAP_values_per_fold = [] #-#-#

        for i, (train_outer_ix, test_outer_ix) in enumerate(zip(ix_training, ix_test)): #-#-#
            print('\n------ Fold Number:',i)
            x_train, x_test = X_train.iloc[train_outer_ix, :], X_train.iloc[test_outer_ix, :]
            y_train, y_test = Y_train[train_outer_ix],   Y_train[test_outer_ix]

            x_train = sc.fit_transform(x_train)
            x_test = sc.transform(x_test)

            svm = SVC(kernel='rbf', C=C_VAL, gamma=R_VAL)
            svm.fit(x_train, y_train)
            print("Accuracy = ", svm.score(X=x_test, y=y_test))

            # Use SHAP to explain predictions
            svm_explainer = shap.KernelExplainer(svm.predict, x_test, feature_names=cols)
            # shap_values = svm_explainer(x_test)
            shap_values = svm_explainer.shap_values(x_test)

            for SHAPs in shap_values:
                SHAP_values_per_fold.append(SHAPs) #-#-#

        len(SHAP_values_per_fold)

        #SUMMARY SHAP PLOT
        new_index = [ix for ix_test_fold in ix_test for ix in ix_test_fold]
        X_train.reindex(new_index)

        plt.figure(figsize=(12, 8))
        shap.summary_plot(np.array(SHAP_values_per_fold), X_train.reindex(new_index), plot_type="bar", max_display=50, show=False, plot_size=(10,8))
        plt.title(f'\nSUMMARY_SHAP_Speed_{sheet_name}_Model_{feat_model_name[counter]}')  # Agrega el título
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'shap_summary_plot_{sheet_name}_{feat_model_name[counter]}.png'))  # Guarda el gráfico con el nombre correspondiente

        #TRAIN DATA SHAP PLOT
        vals              = np.array(SHAP_values_per_fold)
        index             = np.array(new_index).reshape(-1,1)
        shap_values_train = np.append(index,vals, axis = 1)

        plt.figure(figsize=(12, 8))
        shap.summary_plot(np.array(shap_values_train[:, 1:]), X_train.reindex(shap_values_train[:, 0]), plot_type="bar", max_display=50, show=False, plot_size=(10,8))
        plt.title(f'\nTRAIN_SHAP_Speed_{sheet_name}_Model_{feat_model_name[counter]}')  # Agrega el título
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'shap_TRAIN_plot_{sheet_name}_{feat_model_name[counter]}.png'))  # Guarda el gráfico con el nombre correspondiente

        df_cols  = ['Participant']
        df_cols.extend(X_train.reindex(shap_values_train[:, 0]).columns.values.tolist())
        df_Train = pd.DataFrame(shap_values_train, columns=df_cols)

        #TEST DATA SHAP PLOT
        X_train = sc.fit_transform(X_train)
        X_test  = sc.transform(X_test)

        svm = SVC(kernel='rbf', C=C_VAL, gamma=R_VAL)
        svm.fit(X_train, Y_train)
        print("Accuracy = ", svm.score(X=X_test, y=Y_test))

        # Use SHAP to explain predictions
        svm_explainer = shap.KernelExplainer(svm.predict, X_test, feature_names=cols)
        shap_values   = svm_explainer.shap_values(X_test)

        plt.figure(figsize=(12, 8))
        shap.summary_plot(shap_values, feature_names=cols, plot_type = "bar", max_display= 50, show=False, plot_size=(10,8))
        plt.title(f'TEST_SHAP_Speed_{sheet_name}_Model_{feat_model_name[counter]}')
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'SHAP_TEST_plot_{sheet_name}_{feat_model_name[counter]}.png'))  # Guarda el gráfico con el nombre correspondiente

        vals    = np.array(shap_values)
        index   = np.arange(vals.shape[0]).reshape(vals.shape[0],1)
        shap_values_test = np.append(index, vals, axis = 1)
        df_Test = pd.DataFrame(shap_values_test, columns=df_cols)

        df_Train_sorted = df_Train.sort_values(by=df_Train.columns[0])
        df_Train_sorted['Participant'] = df_Train_sorted['Participant'].astype(int)
        df_Train_sorted.set_index('Participant', inplace=True)

        df_Test_sorted  = df_Test.sort_values(by=df_Train.columns[0])
        df_Test_sorted['Participant'] = df_Test_sorted['Participant'].astype(int)
        df_Test_sorted.set_index('Participant', inplace=True)
        # Concatenate df_inf_Train with df_Train_sorted column-wise
        df_combined_Train = pd.concat([df_inf_Train, df_Train_sorted], axis=1)
        # Concatenate df_inf_Test with df_Test_sorted column-wise
        df_combined_Test = pd.concat([df_inf_Test, df_Test_sorted], axis=1)


        Features_SHAP_temp = pd.DataFrame({
                'Index'                                           : features,
                f'Feats_model_{sheet_name}_{feat_model_name[counter]}' : features_value.values.flatten()
            })

        Features_SHAP = pd.concat([Features_SHAP, Features_SHAP_temp], axis=1)
        SHAP_DATA[sheet_name][f'{feat_model_name[counter]}'] = pd.concat([df_combined_Train, df_combined_Test], axis=0)

        print('Actual loop position: ', counter)
        print('Feuture Inx: ', features_indx[counter])
        print('Feature length: ', len(features))
        counter += 1

with pd.ExcelWriter(f'{path_folder}/SHAP_DATA_older.xlsx') as writer:
    for speed in SHAP_DATA:
        for model in SHAP_DATA[speed]:
            SHAP_DATA[speed][model].to_excel(writer, sheet_name=f'{speed}_{model}', index=False)

# with pd.ExcelWriter('Features_from_SHAP.xlsx') as writer:
#     for speed in SHAP_DATA:
#         for model in SHAP_DATA[speed]:
#             Features_SHAP.to_excel(writer, sheet_name=f'Feat_from_ {speed}_{model}', index=True)
Features_SHAP.to_excel(f'{path_folder}/Features_from_SHAP_older.xlsx')

# Read the Excel file
#EXCEL_train = pd.ExcelFile(path_train)
#EXCEL_test = pd.ExcelFile(path_test)
#EXCEL_scores = pd.ExcelFile(path_scores)

# Dictionary to hold the data from each sheet
sheets_data   = {}
SHAP_DATA     = {}
Features_SHAP = pd.DataFrame()

for sheet_name in EXCEL_scores.sheet_names:
    # READ TRAIN SET
    # LEER EL DATA SET DE DATOS PRINCIPAL
    df_train      = pd.read_excel(EXCEL_train, sheet_name=sheet_name)
    df_inf_Train  = df_train.iloc[:,0:4]

    df_Scores     = pd.read_excel(EXCEL_scores, sheet_name=sheet_name)

    SHAP_DATA[sheet_name] = {}

    # model parameters
    C_VAL_list      = list(df_Scores['C_Value'].iloc)
    R_VAL_list      = list(df_Scores['Gamma_Value'].iloc)
    CV_SPLIT_list   = list(df_Scores['CV_split'].iloc)
    RS_list         = list(df_Scores['Random_State'].iloc)
    features_list   = list(df_Scores['Features'].iloc)
    features_indx   = list(df_Scores['Feature_idx'].iloc)
    feat_model_name = list(df_Scores['Name_Model'].iloc)
    df_participants = df_train.iloc[:,0:4]

    # THE FIRST AND LEAST FEATURES
    a = 'Pelv_Angle_Y_MAX_SW'
    b = 'Hip_Angle_Z_OHS'

    # DEFINIR EL CAMBIO DE VARIABLE EN LA CLASIFICACIÓN OBJETIVA -y
    df_train["Group"] = df_train["Group"].replace({'Control': 0, 'Flatfoot': 1})

    # Define the different segments from dataset to be used.
    segments_train = {
        'All_Data' : df_train.loc[:, a:b]
    }

    # saving the target variables into `y` variable.
    Y_train = df_train.loc[:, "Group"].values

    # READ TEST SET
    df_test_p   = pd.read_excel(EXCEL_test, sheet_name=sheet_name)
    df_inf_Test = df_test_p.iloc[:,0:4]
    df_test = df_test_p.drop(["Participant", "Index", "Side"], axis=1)
    df_test["Group"] = df_test["Group"].replace({'Control': 0, 'Flatfoot': 1})

    segments_test = {
        'All_Data' : df_test.loc[:, a : b]
    }

    Y_test  = df_test.loc[:, "Group"].values

    # CREATE THE FOLDER WHERE THE SHAP'S PLOTS WILL BE SAVED
    output_dir = f'{path_folder}/SHAPS_TRAIN_and_TEST_Plots'
    os.makedirs(output_dir, exist_ok=True)

    sc = StandardScaler()
    counter = 0

    for features_str in features_list:
        # model parameters
        C_VAL = C_VAL_list[counter]
        R_VAL = R_VAL_list[counter]
        CV_SPLIT     = CV_SPLIT_list[counter]
        RANDOM_STATE = RS_list[counter]

        # GET NAME MODEL
        df_name = f'Model_{feat_model_name[counter]}'

        # TRANSFORM THE DATA ON THE CORRECT FORMAT
        features = features_str.strip("[]").split()
        features = np.array(features)
        sfs_feature_names = [int(feature) for feature in features]

        print("Number of Features selected: ", len(sfs_feature_names))
        print("df.shape = ", segments_train['All_Data'].iloc[:, sfs_feature_names].shape)

        segment_train = segments_train['All_Data'].iloc[:, sfs_feature_names]
        segment_test  = segments_test['All_Data'].iloc[:, sfs_feature_names]
        cols = segment_train.columns.values.tolist()
        print(len(cols))

        temp = pd.DataFrame(segments_train["All_Data"].keys().to_numpy(), columns=["FeatureNames"])
        features_value = temp.iloc[sfs_feature_names]

        kfold_cv = StratifiedKFold(n_splits = CV_SPLIT, shuffle=True, random_state = RANDOM_STATE)
        X_train  = segment_train
        X_test   = segment_test

        ix_training, ix_test = [], []
        # Loop through each fold and append the training & test indices to the empty lists above
        for fold in kfold_cv.split(X_train, Y_train):
            ix_training.append(fold[0]), ix_test.append(fold[1])
            print(len(fold[1]))

        len(ix_training)
        SHAP_values_per_fold = [] #-#-#

        for i, (train_outer_ix, test_outer_ix) in enumerate(zip(ix_training, ix_test)): #-#-#
            print('\n------ Fold Number:',i)
            x_train, x_test = X_train.iloc[train_outer_ix, :], X_train.iloc[test_outer_ix, :]
            y_train, y_test = Y_train[train_outer_ix],   Y_train[test_outer_ix]

            x_train = sc.fit_transform(x_train)
            x_test = sc.transform(x_test)

            svm = SVC(kernel='rbf', C=C_VAL, gamma=R_VAL)
            svm.fit(x_train, y_train)
            print("Accuracy = ", svm.score(X=x_test, y=y_test))

            # Use SHAP to explain predictions
            svm_explainer = shap.KernelExplainer(svm.predict, x_test, feature_names=cols)
            # shap_values = svm_explainer(x_test)
            shap_values = svm_explainer.shap_values(x_test)

            for SHAPs in shap_values:
                SHAP_values_per_fold.append(SHAPs) #-#-#

        len(SHAP_values_per_fold)

        #SUMMARY SHAP PLOT
        new_index = [ix for ix_test_fold in ix_test for ix in ix_test_fold]
        X_train.reindex(new_index)

        plt.figure(figsize=(12, 8))
        shap.summary_plot(np.array(SHAP_values_per_fold), X_train.reindex(new_index), plot_type="bar", max_display=50, show=False, plot_size=(10,8))
        plt.title(f'\nSUMMARY_SHAP_Speed_{sheet_name}_Model_{feat_model_name[counter]}')  # Agrega el título
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'shap_summary_plot_{sheet_name}_{feat_model_name[counter]}.png'))  # Guarda el gráfico con el nombre correspondiente

        #TRAIN DATA SHAP PLOT
        vals              = np.array(SHAP_values_per_fold)
        index             = np.array(new_index).reshape(-1,1)
        shap_values_train = np.append(index,vals, axis = 1)

        plt.figure(figsize=(12, 8))
        shap.summary_plot(np.array(shap_values_train[:, 1:]), X_train.reindex(shap_values_train[:, 0]), plot_type="bar", max_display=50, show=False, plot_size=(10,8))
        plt.title(f'\nTRAIN_SHAP_Speed_{sheet_name}_Model_{feat_model_name[counter]}')  # Agrega el título
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'shap_TRAIN_plot_{sheet_name}_{feat_model_name[counter]}.png'))  # Guarda el gráfico con el nombre correspondiente

        df_cols  = ['Participant']
        df_cols.extend(X_train.reindex(shap_values_train[:, 0]).columns.values.tolist())
        df_Train = pd.DataFrame(shap_values_train, columns=df_cols)

        #TEST DATA SHAP PLOT
        X_train = sc.fit_transform(X_train)
        X_test  = sc.transform(X_test)

        svm = SVC(kernel='rbf', C=C_VAL, gamma=R_VAL)
        svm.fit(X_train, Y_train)
        print("Accuracy = ", svm.score(X=X_test, y=Y_test))

        # Use SHAP to explain predictions
        svm_explainer = shap.KernelExplainer(svm.predict, X_test, feature_names=cols)
        shap_values   = svm_explainer.shap_values(X_test)

        plt.figure(figsize=(12, 8))
        shap.summary_plot(shap_values, feature_names=cols, plot_type = "bar", max_display= 50, show=False, plot_size=(10,8))
        plt.title(f'TEST_SHAP_Speed_{sheet_name}_Model_{feat_model_name[counter]}')
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'SHAP_TEST_plot_{sheet_name}_{feat_model_name[counter]}.png'))  # Guarda el gráfico con el nombre correspondiente

        vals    = np.array(shap_values)
        index   = np.arange(vals.shape[0]).reshape(vals.shape[0],1)
        shap_values_test = np.append(index, vals, axis = 1)
        df_Test = pd.DataFrame(shap_values_test, columns=df_cols)

        df_Train_sorted = df_Train.sort_values(by=df_Train.columns[0])
        df_Train_sorted['Participant'] = df_Train_sorted['Participant'].astype(int)
        df_Train_sorted.set_index('Participant', inplace=True)

        df_Test_sorted  = df_Test.sort_values(by=df_Train.columns[0])
        df_Test_sorted['Participant'] = df_Test_sorted['Participant'].astype(int)
        df_Test_sorted.set_index('Participant', inplace=True)
        # Concatenate df_inf_Train with df_Train_sorted column-wise
        df_combined_Train = pd.concat([df_inf_Train, df_Train_sorted], axis=1)
        # Concatenate df_inf_Test with df_Test_sorted column-wise
        df_combined_Test = pd.concat([df_inf_Test, df_Test_sorted], axis=1)


        Features_SHAP_temp = pd.DataFrame({
                'Index'                                           : features,
                f'Feats_model_{sheet_name}_{feat_model_name[counter]}' : features_value.values.flatten()
            })

        Features_SHAP = pd.concat([Features_SHAP, Features_SHAP_temp], axis=1)
        SHAP_DATA[sheet_name][f'{feat_model_name[counter]}'] = pd.concat([df_combined_Train, df_combined_Test], axis=0)

        print('Actual loop position: ', counter)
        print('Feuture Inx: ', features_indx[counter])
        print('Feature length: ', len(features))
        counter += 1

with pd.ExcelWriter(f'{path_folder}/SHAP_DATA_older.xlsx') as writer:
    for speed in SHAP_DATA:
        for model in SHAP_DATA[speed]:
            SHAP_DATA[speed][model].to_excel(writer, sheet_name=f'{speed}_{model}', index=False)

# with pd.ExcelWriter('Features_from_SHAP.xlsx') as writer:
#     for speed in SHAP_DATA:
#         for model in SHAP_DATA[speed]:
#             Features_SHAP.to_excel(writer, sheet_name=f'Feat_from_ {speed}_{model}', index=True)
Features_SHAP.to_excel(f'{path_folder}/Features_from_SHAP_older.xlsx')
```
