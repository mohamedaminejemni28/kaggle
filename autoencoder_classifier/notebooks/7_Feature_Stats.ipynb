# Obtenain stats from unique features from the chosen models

## Import libraries

```python
import numpy as np
import pandas as pd
```

## Load Excel File with the stats

```python
path_stats = 'Sample Data/Multivars_stats_older.xlsx'

df_stats = pd.read_excel(path_stats, header=None)
df_stats = df_stats.transpose()
df_stats.columns = df_stats.loc[0][:]
df_stats = df_stats.loc[1:][:]
df_stats.head()
```

## Load the excel file with the unique features

```python
path_invariant = 'Sample Data/Unique_features_older.xlsx'
#path_variant   = 'Unique_features_variant.xlsx'

Excel_invariant = pd.ExcelFile(path_invariant)
#Excel_variant   = pd.ExcelFile(path_variant)
```

```python
unique_ft_stats_invariant = {}
#unique_ft_stats_variant = {}
```

```python
stats_columns = df_stats.iloc[:,0]
stats_columns
```

```python
for sheet_name in Excel_invariant.sheet_names:
    df_unique_ft = pd.read_excel(Excel_invariant, sheet_name=sheet_name)
    unique_ft = df_unique_ft['Feature_Name']
    temp_df_stats = df_stats[unique_ft]
    unique_ft_stats_invariant[sheet_name] = temp_df_stats.set_index(stats_columns).transpose()
```

```python
for sheet_name in Excel_variant.sheet_names:
    df_unique_ft = pd.read_excel(Excel_variant, sheet_name=sheet_name)
    unique_ft     = df_unique_ft['Feature_Name']
    temp_df_stats = df_stats[unique_ft]
    unique_ft_stats_variant[sheet_name] = temp_df_stats.set_index(stats_columns).transpose()
```

```python
# Save all filtered sheets to a new Excel file
with pd.ExcelWriter('Feature_stats_older.xlsx') as writer:
    for sheet_name, df_filtered in unique_ft_stats_invariant.items():
        df_filtered.to_excel(writer, sheet_name=sheet_name, index=True)
```

```python
# Save all filtered sheets to a new Excel file
with pd.ExcelWriter('Feature_stats_variant.xlsx') as writer:
    for sheet_name, df_filtered in unique_ft_stats_variant.items():
        df_filtered.to_excel(writer, sheet_name=sheet_name, index=True)
```
