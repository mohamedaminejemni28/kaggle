# Split main data set for obtain TEST and TRAIN sets

# Import libraries

```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

## Load the datasets with non-highly correlated fetaures

```python
file_invariant_set = "ALL FEATURES OLDER - Pes Planus and Control NH Correlated features JUNE 25.xlsx"
Ex_invariant = pd.ExcelFile(file_invariant_set)
```

## Itarate to obtain the split data_frames for invariant set

```python
# Create diccionary will save the split sets
df_invariant_train_set = {}
df_invariant_test_set = {}

# Iterate through each sheet in the Excel file
for sheet_name in Ex_invariant.sheet_names:
    print('Current page: ',sheet_name)
    # Load the sheet into a DataFrame
    df = pd.read_excel(file_invariant_set, sheet_name=sheet_name)
    print('ALL Columns: ', df.shape)

    # Identify all data and objective variable 'Y'
    X=df
    y=df.loc[:, 'Group']

    # Split SETS
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.20, random_state=0)
    print('X Train set: ', X_train.shape)
    print('X Test set: ', X_test.shape)

    # Save sets into the dictionary
    df_invariant_train_set[sheet_name] = pd.DataFrame(X_train)
    df_invariant_test_set[sheet_name] = pd.DataFrame(X_test)
```

## Save Train and Test sets into new Excel Files

```python
# Save all sheets to a new Excel file
output_path_train = 'ALL FEATURES OLDER - Pes Planus and Control JUNE 25 Train.xlsx'
output_path_test = 'ALL FEATURES OLDER - Pes Planus and Control JUNE 25 Test.xlsx'
```

### Invariant TRAIN SET

```python
# Invariant train file
with pd.ExcelWriter(output_path_train) as writer:
    for sheet_name, X_train in df_invariant_train_set.items():
        X_train.to_excel(writer, sheet_name=sheet_name, index=False)
```

### Invariant TEST SET

```python
# Invariant TEST file
with pd.ExcelWriter(output_path_test) as writer:
    for sheet_name, X_test in df_invariant_test_set.items():
        X_test.to_excel(writer, sheet_name=sheet_name, index=False)
```

## Itarate to obtain the split data_frames for variant set

```python
# Invariant train file
with pd.ExcelWriter(output_path_train) as writer:
    for sheet_name, X_train in df_invariant_train_set.items():
        X_train.to_excel(writer, sheet_name=sheet_name, index=False)
# Invariant train file
with pd.ExcelWriter(output_path_test) as writer:
    for sheet_name, X_train in df_invariant_test_set.items():
        X_train.to_excel(writer, sheet_name=sheet_name, index=False)
```
