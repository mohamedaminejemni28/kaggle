## Import libraries


```python
import numpy as np
import pandas as pd
```

## Load Excel File with the features correlated list

```python
file_name = "Highly correlated features.xlsx"
df_JA_TS = pd.read_excel(file_name, header=None)
df_JA_TS.rename(columns={0: "Feature_names"}, inplace=True)

print(df_JA_TS.head())
```

### Import variant AND invariant Data sets

```python
#Names from the Excel files
file_name_invariant_sp = 'ALL FEATURES OLDER - Pes Planus and Control JUNE 25.xlsx'
```

```python
#Load EXCEL_Files
EXCEL_invariant_sp = pd.ExcelFile(file_name_invariant_sp)
```

### Validate that all features to drop exist on the differents sheet names from the EXCEL files


```python
for sheet_name in EXCEL_invariant_sp.sheet_names:
    df = pd.read_excel(file_name_invariant_sp, sheet_name=sheet_name)
    columns_to_drop = df_JA_TS['Feature_names'].tolist()
    missing_columns = [col for col in columns_to_drop if col not in df.columns]
    print("Missing columns:", missing_columns, "\nIn the sheet name called: ", sheet_name, "\nFrome the invariant Excel file called: ", file_name_invariant_sp)
```

### Filter the dataset to include only the Joint Anlges and Temporal-Spatial features

```python
filter_invariant_sheets = {}
output_path = "ALL FEATURES OLDER - Pes Planus and Control NH Correlated feattures JUNE 25.xlsx"

# Iterate through each sheet in the Excel file
for sheet_name in EXCEL_invariant_sp.sheet_names:
    # Load the sheet into a DataFrame
    df = pd.read_excel(file_name_invariant_sp, sheet_name=sheet_name)
    # Filter the DataFrame to keep index columns and cleaned features
    df_filtered = df.drop(columns=columns_to_drop, errors="ignore")
    # Store the filtered DataFrame in the dictionary
    filter_invariant_sheets[sheet_name] = df_filtered

# Save all filtered sheets to a new Excel file
with pd.ExcelWriter(output_path) as writer:
    for sheet_name, df_filtered in filter_invariant_sheets.items():
        df_filtered.to_excel(writer, sheet_name=sheet_name, index=False)
```
