## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Create variables with the all the Joint Angle and Temporal-Spatial Feauture Names

### Import variables names from a excel spreadsheet file 

```python
file_neme = "ISB Abstract - Joint Angles and TS Variable Names.xlsx"
df_JA_TS = pd.read_excel(file_neame)
print(df_JA_TS.head())
```

### Import variant AND invariant Data sets

```python
#Names from the Excel files
file_name_variant_sp = '2024 ALL Collected Speeds Young and Older Adults ALL Features - FINAL AUGUST 20.xlsx'
file_name_invariant_sp = '2024 ALL Post-Processing Speeds Young and Older Adults ALL Features - FINAL AUGUST 20.xlsx'
```

```python
#Load data_frames
df_variant_sp = pd.read_excel(file_name_variant_sp)
print(df_variant_sp.head())
```

```python
df_invariant_sp = pd.read_excel(file_name_invariant_sp)
print(df_invariant_sp.head())
```

## Create a new DataFrame with only the desired features

```python
# Define the columns to keep (first four columns)
index_columns_variant_sp = df_variant_sp.columns[:4].tolist()  # Adjust this if your index columns vary
print(index_columns_variant_sp)
```

```python
# Define the columns to keep (first four columns)
index_columns_invariant_sp = df_invariant_sp.columns[:4].tolist()  # Adjust this if your index columns vary
print(index_columns_invariant_sp)
```

```python
# Create a list with the focused features
# Combine the two columns into a single list and remove NaN values
combined_features = pd.concat([df_JA_TS['Joint Angle Features'], 
                                df_JA_TS['Temporal-Spatial Features']])

# Drop NaN values
JA_TS_ft = combined_features.dropna().tolist()
JA_TS_ft_variant = index_columns_variant_sp + JA_TS_ft 
JA_TS_ft_invariant = index_columns_invariant_sp + JA_TS_ft
print(JA_TS_ft[0:10])
```

### Filter the dataset to include only the Joint Anlges and Temporal-Spatial features

#### Filter variant data set

```python
df_variant_filtered = df_variant_sp.loc[:, df_variant_sp.columns.isin(JA_TS_ft_variant)]
df_variant_filtered.head()
```

#### Filter Variant data set

```python
df_invariant_filtered = df_invariant_sp.loc[:, df_invariant_sp.columns.isin(JA_TS_ft_invariant)]
df_invariant_filtered.head()
```

# Save the filtered datasets

```python
df_variant_filtered.to_excel("2024 ALL Collected Speeds Young and Older Adults JA and TS features - FINAL AUGUST 20.xlsx", index=False)
df_invariant_filtered.to_excel("2024 ALL Post-Processing Speeds Young and Older Adults JA and TS features - FINAL AUGUST 20.xlsx", index=False)
```
