# Unique Features Code

```python
"""
Created on Sat Jul 13 13:02:44 2024

@author: roqui
"""
import pandas            as pd  # for reading the .csv file and related operations
import numpy             as np  # for working with arrays (multi-dimensional)

path_folder = f'Sample Data'
#path_folder = f'Variant Speeds Young and Older'
path_train  = f'{path_folder}/ALL FEATURES OLDER - Pes Planus and Control JUNE 25 Train.xlsx'
path_test   = f'{path_folder}/ALL FEATURES OLDER - Pes Planus and Control JUNE 25 Test.xlsx'
path_scores = f'{path_folder}/Scores_RBF_older.xlsx'

EXCEL_scores = pd.ExcelFile(path_scores)
EXCEL_train  = pd.ExcelFile(path_train)

unique_ft = {}

for sheet_name in EXCEL_scores.sheet_names:

    df_scores   = pd.read_excel(EXCEL_scores, sheet_name=sheet_name)
    df_features = pd.read_excel(EXCEL_train, sheet_name=sheet_name)

    # GET AS A LIST THE TOP 2 SFS MODELS THAT GET A BETTER PERFORMANCE
    sigificant_sfs_models = list(df_scores['Features'].iloc)
    all_features          = df_features.columns[4:]
    unique_features = []

    for i in sigificant_sfs_models:

        list_values = i.strip('[]').split()
        feat_ind    = []
        for j in list_values:
            feat_ind.append(int(j))
        unique_features.extend(set(feat_ind))

    unique_features = np.unique(unique_features)
    name_features   = all_features[unique_features]

    # Create a DataFrame with indices and feature names
    unique_feats_df = pd.DataFrame({
        'Feature_Index': unique_features,
        'Feature_Name': name_features.values
    })

    # Save the DataFrame to an Excel file
    unique_ft[sheet_name] = unique_feats_df

with pd.ExcelWriter('Sample Data/Unique_features_older.xlsx') as writer:
    for sheet_name, df_results in unique_ft.items():
        df_results.to_excel(writer, sheet_name=sheet_name, index=True)
```
